```python
# I.) Scrape The Data From Lazada
import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup

def getUrl(key):
    key = key.replace(' ','+')
    temp = 'https://www.lazada.com.my/tag/{}'
    url = temp.format(key)
    return url

def get_products(card):
    # Fetching product_name
    try:
        product_name = card.find('div','RfADt').text
    except:
        product_name = 'Not Mentioned'
# -----------------------------------------------------------
    # Fetching product_price
    try:
        product_price = card.find('span','ooOxS').text
    except:
        product_price = 'Not Mentioned'
# -----------------------------------------------------------
    # Fetching product_desc
    try:
        product_desc = card.find('span','IcOsH').text
    except:
        product_desc = 'Not Mentioned'
# -----------------------------------------------------------
    # Fetching product_img_url
    try:
        product_img_url = card.find('div','picture-wrapper').find('img')
        product_img_url = product_img_url['src']
    except:
        product_img_url = 'Not Mentioned'
# -----------------------------------------------------------
    # Fetching product_buy_link
    try:
        product_buy_link = card.find('div','RfADt').find('a')
        product_buy_link = 'https:'+ product_buy_link['href'] 
    except:
        product_buy_link = 'Not Mentioned'
    
    all_product_info = (product_name,product_price,product_desc,product_img_url,product_buy_link)
    return all_product_info

def main(product):
    all_products = []
    url = getUrl('earbuds')
    driver = webdriver.Firefox()
    driver.get(url)
    time.sleep(8)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    cards = soup.find_all('div','Bm3ON')
    for i in cards:
        temp = get_products(i)
        all_products.append(temp)
    col = ['product_name','product_price','product_desc','product_img_url','product_buy_link']
    lazada_product_data = pd.DataFrame(all_products,columns=col)
    lazada_product_data.to_csv('D:\\Acadamics\\Internship\\iGurus\\Projects\\FlipCart_lazada_webscrapping\\lazada_product_data.csv',encoding='utf-8')
```

```python
main('earbuds')
```
